```python
import pandas as pd

df = pd.read_excel('pt_troubleshooting.xlsx')
df.head()
```

```python
problem = list(df['problem'].unique())
equipament = list(df['equipament'].unique())
print(len(problem))
print(len(equipament))
```

```python
problems = []

for i in problem:
    i = i.lower().strip()
    i = i.split('.')
    if type(i) == list:
        for j in i:
            j = j.strip()
            if '(continuação)' not in j:
                problems.append(j)
    else:
        i = i.strip()
        if '(continuação)' not in i:
            problems.append(i)

len(problems)
```

```python
def escrever_em_arquivo(lista, nome_arquivo):
    with open(nome_arquivo, 'w',encoding='utf-8') as arquivo:
        count = 0
        for elemento in lista:
            if elemento != '':
                count += 1
                arquivo.write(f'{count} - {str(elemento)} \n')

escrever_em_arquivo(problems,'..\lista de problemas.txt')
```
